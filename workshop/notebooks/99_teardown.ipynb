# 99 - Teardown

**Destroy all live-provisioned resources in reverse dependency order.**

Order: postgres -> object_storage

After running this notebook, run `tofu destroy` in `workshop/infrastructure/` from your local machine to destroy the Jupyter instance itself.

````python
import subprocess
import os
from dotenv import load_dotenv

load_dotenv()
project_suffix = os.environ.get("PROJECT_SUFFIX", "unknown")
print(f"Tearing down resources for student: {project_suffix}")
````

````python
# 2. Destroy PostgreSQL
print("=== Destroying PostgreSQL ===")
result = subprocess.run(
    ["tofu", "destroy", "-auto-approve"],
    cwd="../iac_snippets/postgres",
    capture_output=True,
    text=True,
)
print(result.stdout[-300:] if len(result.stdout) > 300 else result.stdout)
if result.returncode != 0:
    print("STDERR:", result.stderr[-300:])
else:
    print("PostgreSQL destroyed.")
````

````python
# 3. Destroy Object Storage
print("=== Destroying Object Storage ===")
result = subprocess.run(
    ["tofu", "destroy", "-auto-approve"],
    cwd="../iac_snippets/object_storage",
    capture_output=True,
    text=True,
)
print(result.stdout[-300:] if len(result.stdout) > 300 else result.stdout)
if result.returncode != 0:
    print("STDERR:", result.stderr[-300:])
else:
    print("Object Storage destroyed.")
````

````python
# Verification: check for lingering resources
print("=== Verification ===")
print("If all destroys succeeded, no workshop resources should remain.")
print()
print("IMPORTANT: To destroy the Jupyter instance itself, run from your local machine:")
print("  cd workshop/infrastructure && tofu destroy")
````

## Done!

All live-provisioned resources have been destroyed.
