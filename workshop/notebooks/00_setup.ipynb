### Environment Setup

This cell loads your Scaleway credentials from the `.env` file and validates that all required variables are present.


```python
import os
from dotenv import load_dotenv

load_dotenv()

# Check required environment variables
required_vars = [
    "SCW_ACCESS_KEY",
    "SCW_SECRET_KEY",
    "SCW_DEFAULT_ORGANIZATION_ID",
    "SCW_DEFAULT_PROJECT_ID",
    "SCW_DEFAULT_REGION",
    "SCW_DEFAULT_ZONE",
    "PROJECT_SUFFIX",
]

missing = [v for v in required_vars if not os.environ.get(v)]
if missing:
    print(f"ERROR: Missing environment variables: {', '.join(missing)}")
    print("Check /home/jupyter/.env")
else:
    print("All environment variables loaded.")
    print(f"Project suffix: {os.environ['PROJECT_SUFFIX']}")
    print(f"Region: {os.environ['SCW_DEFAULT_REGION']}")
```

### OpenTofu - Infrastructure as Code

[OpenTofu](https://opentofu.org/) is an open-source alternative to Terraform, used to define and provision cloud infrastructure using code.  
Instead of clicking through cloud consoles, you describe your resources (servers, networks, storage) in `.tf` files - and OpenTofu creates them automatically.  
This cell verifies that OpenTofu is installed and ready to use in your environment.

```python
# Check OpenTofu
import subprocess

result = subprocess.run(["tofu", "--version"], capture_output=True, text=True)
print(result.stdout.strip())
```

###  Scaleway CLI

The Scaleway CLI (`scw`) is a command-line tool for managing Scaleway cloud resources directly from the terminal.  
This cell confirms it is installed and displays the current version.

```python
# Check Scaleway CLI
result = subprocess.run(["scw", "version"], capture_output=True, text=True)
print(result.stdout.strip())
```
