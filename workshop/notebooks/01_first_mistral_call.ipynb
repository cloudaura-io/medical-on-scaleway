###  Connect to Scaleway Generative APIs

This cell initializes the connection to **Scaleway's Generative APIs** - the endpoint that gives you access to hosted AI models like Mistral.

We use the `OpenAI` Python client, which is compatible with Scaleway's API out of the box.  
Your secret key is loaded from the `.env` file - never hardcoded.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

client = OpenAI(
    base_url="https://api.scaleway.ai/v1",
    api_key=os.environ["SCW_SECRET_KEY"],
)

print("Connected to Scaleway Generative APIs.")
```

### Understanding Temperature

**Temperature** controls how creative (or predictable) the model's responses are.

- **Low temperature (≈ 0)** → model is focused and consistent - nearly the same answer every time
- **High temperature (≈ 1+)** → model is more creative and varied - different answers each run

This cell runs the **same medical prompt 3 times** with a very low temperature (`0.01`) to demonstrate how the model behaves when determinism matters - critical in clinical contexts where consistency is expected.

```python
prompt = "In one sentence, what should a patient avoid eating while taking KETOCONAZOLE?"
```

```python
# Temperature = 0.01 (near-deterministic): run the same prompt 3 times
print("=== Temperature 0.01 (low variance) ===\n")
for i in range(3):
    response = client.chat.completions.create(
        model="mistral-small-3.2-24b-instruct-2506",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.01,
        max_tokens=100,
    )
    print(f"Run {i + 1}: {response.choices[0].message.content}")
    print()
```

###  High Temperature - Creative but Unpredictable

Now we run the **same prompt** with `temperature=1.0`.

Notice how each response differs in wording, structure, and emphasis - even though the question is identical.  
In healthcare, this variability can be dangerous: the model may omit critical details or phrase risks inconsistently.

> ⚠️ **Key takeaway:** For medical applications, always use low temperature to ensure reliable, reproducible outputs.

```python
# Temperature = 1.0 (high variance): same prompt, observe the differences
print("=== Temperature 1.0 (high variance) ===\n")
for i in range(3):
    response = client.chat.completions.create(
        model="mistral-small-3.2-24b-instruct-2506",
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
        max_tokens=100,
    )
    print(f"Run {i + 1}: {response.choices[0].message.content}")
    print()
```

### 🔍 Key Observations

- **Temperature 0.1** → responses are nearly identical across runs - the model behaves deterministically
- **Temperature 1.0** → responses vary in wording and sometimes in substance - randomness increases with each token sampled
- In **both cases**, the model does not cite any specific FDA label or study - the answer is **ungrounded**
