# 01 - First Mistral API Call

**Learning objective:** Make your first call to Scaleway Generative APIs with Mistral Small 3.2.

We will:
1. Connect to the Scaleway Generative APIs endpoint
2. Send a medical prompt to Mistral Small 3.2
3. Observe temperature variance across multiple calls
4. Preview what a ReAct agent will do differently in notebook 06

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

client = OpenAI(
    base_url="https://api.scaleway.ai/v1",
    api_key=os.environ["SCW_SECRET_KEY"],
)

print("Connected to Scaleway Generative APIs.")
```

```python
# Temperature = 0.1 (near-deterministic): run the same prompt 3 times
prompt = "In one sentence, what is the main risk of combining warfarin with NSAIDs?"

print("=== Temperature 0.1 (low variance) ===\n")
for i in range(3):
    response = client.chat.completions.create(
        model="mistral-small-3.2-24b-instruct-2506",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.1,
        max_tokens=100,
    )
    print(f"Run {i + 1}: {response.choices[0].message.content}")
    print()
```

```python
# Temperature = 1.0 (high variance): same prompt, observe the differences
print("=== Temperature 1.0 (high variance) ===\n")
for i in range(3):
    response = client.chat.completions.create(
        model="mistral-small-3.2-24b-instruct-2506",
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
        max_tokens=100,
    )
    print(f"Run {i + 1}: {response.choices[0].message.content}")
    print()
```

## Key observations

- At **temperature 0.1**, responses are nearly identical across runs. The model is deterministic.
- At **temperature 1.0**, responses vary in wording and sometimes in substance. Higher temperature increases randomness in token sampling.
- In both cases, the model does not cite a specific FDA label or study. The answer is **ungrounded**.
- In notebook 06, we build a ReAct agent that grounds every claim in a specific FDA label section with a verifiable citation.
