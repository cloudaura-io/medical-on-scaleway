# 00 - Setup & Environment Check

**Learning objective:** Verify your workshop environment is ready.

This notebook checks that:
1. Environment variables (Scaleway credentials) are loaded
2. JupyterLab is running correctly
3. OpenTofu and Scaleway CLI are installed
4. Required Python packages are available

```python
import os
from dotenv import load_dotenv

load_dotenv()

# Check required environment variables
required_vars = [
    "SCW_ACCESS_KEY",
    "SCW_SECRET_KEY",
    "SCW_DEFAULT_ORGANIZATION_ID",
    "SCW_DEFAULT_PROJECT_ID",
    "SCW_DEFAULT_REGION",
    "SCW_DEFAULT_ZONE",
    "PROJECT_SUFFIX",
]

missing = [v for v in required_vars if not os.environ.get(v)]
if missing:
    print(f"ERROR: Missing environment variables: {', '.join(missing)}")
    print("Check /home/jupyter/.env")
else:
    print("All environment variables loaded.")
    print(f"Project suffix: {os.environ['PROJECT_SUFFIX']}")
    print(f"Region: {os.environ['SCW_DEFAULT_REGION']}")
```

```python
# Check OpenTofu
import subprocess

result = subprocess.run(["tofu", "--version"], capture_output=True, text=True)
print(result.stdout.strip())
```

```python
# Check Scaleway CLI
result = subprocess.run(["scw", "version"], capture_output=True, text=True)
print(result.stdout.strip())
```

```python
# Check Python packages
import openai
import psycopg
import boto3
import gradio
import pandas

print(f"openai:   {openai.__version__}")
print(f"psycopg:  {psycopg.__version__}")
print(f"boto3:    {boto3.__version__}")
print(f"gradio:   {gradio.__version__}")
print(f"pandas:   {pandas.__version__}")
```
